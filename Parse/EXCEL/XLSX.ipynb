```python
from xlrd import open_workbook
import os
```

```python
files_list = os.listdir('YourBoardingPassDotAero')
```

```python
headers_list = ['FamilyStatus',
                'FullName',
                'DepartureDate',
                'DepartureTime',
                'DepartureCity',
                'DepartureAirport',
                'ArrivalCity',
                'ArrivalAirport',
                'E-ticket',
                'Class',
                'Flight',
                'BonusProgram',
                'BookingCode',
                'CodeShare',
                'Sequence'
               ]

def list_to_string(list):
    res_string = ''
    for item in list:
        res_string += (str(item) + ';')
    return res_string[:-1]

res_list = []
res_list.append(list_to_string(headers_list))

len(res_list)
```

```python
coord_seq_dict = {
                'FamilyStatus':['0','2'],
                'FullName':['1','2'],
                'DepartureDate':['0','8'],
                'DepartureTime':['2','8'],
                'DepartureCity':['3','4'],
                'DepartureAirport':['3','6'],
                'ArrivalCity':['7','4'],
                'ArrivalAirport':['7','6'],
                'E-ticket':['4','12'],
                'Class':['7','2'],
                'Flight':['0','4'],
                'BonusProgram':['5','2'],
                'BookingCode':['1','12'],
                'CodeShare':['4','8'],
                'Sequence':['7','0']
                }
```

```python
count_file = 0
for file in files_list:
    print('File: {} from {}'.format(count_file, len(files_list) + 1))
    path = 'YourBoardingPassDotAero/' + file
    wb = open_workbook(path)
    for sheet in wb.sheets():
        sheet_res_list = []
        for key in coord_seq_dict.keys():
            cur_coord_row = coord_seq_dict[key][1]
            cur_coord_col = coord_seq_dict[key][0]
            sheet_res_list.append(sheet.cell(int(cur_coord_row), int(cur_coord_col)).value)
        res_list.append(list_to_string(sheet_res_list))
        
    count_file += 1
```

```python
len(res_list)
```

```python
def to_csv(csv_list):
    with open('excel_all.csv', 'w') as f:
        for line in csv_list:
            f.write(line + '\n')
            
to_csv(res_list)
```
